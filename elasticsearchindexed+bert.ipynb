```python
# Import the necessary libraries
%time
from elasticsearch.helpers import bulk
from elasticsearch import RequestError
import csv


def elaster():
    with  open('tobeindexed.csv') as csv_file:
        reader = csv.reader(csv_file)

        for i,line in enumerate(reader):
          document= {
              'self_link': line[1],
              'title' : line[2],
              'abstract': line[3],	
              'publication_number' : line[4],
              'date': line[5],
              'inventors' : line[6],
          }
          yield {
              '_index': 'patents',
                '_id': i,
                '_source': document,
          }

try:
    es.indices.create('patents')
    bulk(es, elaster())
except RequestError:
    print("Not recreating index that already exists")
```

```python
hits = es.search(index='patents', body={"query": {"match": {"title": "current"}}})

for hit in hits['hits']['hits']:
    print(f"Score: {hit['_score']:.2f}, Title: {hit['_source']['title']}")
```

```python
import numpy as np
import torch
import os
import pandas as pd
import faiss
import time
import pickle
from pathlib import Path
from sentence_transformers import SentenceTransformer
from transformers import *
```

```python
df=pd.read_csv('/content/drive/MyDrive/tobeindexed.csv')
```

```python
df
```

```python
import re
pattern = r'\d+'
extract_numeric = lambda row: re.search(pattern, row['publication_number']).group()
df['numeric_value'] = df.apply(extract_numeric, axis=1)
```

```python
df=df.iloc[:25000]
```

```python
new_df = df.drop(df.columns[[0,4,6]], axis=1)
```

```python
new_df=new_df.iloc[:25000]
```

```python
new_df['inventor'] = new_df['inventors'].apply(lambda x: x[1:-1])
```

```python
new_df['inventor']
```

```python
new_df
```

```python
new_df['combined'] = new_df[['abstract','title','inventor']].apply(lambda x: ' [SEP] '.join(x.astype(str)), axis=1)
```

```python
new_df
```

```python
#model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

if torch.cuda.is_available(): 
    model = model.to(torch.device("cuda"))
print(model.device)
```

```python
embeddings = model.encode(new_df.publication_number.to_list(), show_progress_bar=True)
```

```python
embeddings1= model.encode(new_df.combined.to_list(), show_progress_bar=True)
```

```python
combined_embeddings = 0.7 * embeddings + 0.3 * embeddings1
```

```python
print(combined_embeddings.shape)
```

```python
embeddings = np.array([_ for _ in combined_embeddings]).astype("float32")
index = faiss.IndexFlatL2(embeddings.shape[1])
index = faiss.IndexIDMap(index)
index.add_with_ids(embeddings, df.numeric_value.values)
```

```python
def indx(df, I):
    I=I.flatten().tolist()
    return [df.loc[df.numeric_value == str(idx)].to_dict(orient='records') for idx in I]
    
def search(query, model, index, num_results):
  vector = model.encode(list(query))
  D, I = index.search(np.array(vector).astype("float32"), k=num_results)
  return D, I
```

```python
query="WO2005033978A1"
D, I = search([query], model, index, num_results=10)
print(f'Distance: {D.flatten().tolist()}\n\Publication_number: {I.flatten().tolist()}')
```

```python
res=indx(df, I)
for i in res:
  print(i[0])
```
